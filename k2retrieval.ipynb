```python
#imports
import lightkurve as lk
import pandas as pd
from tqdm import tqdm
from lightkurve.utils import LightkurveError
import numpy as np
```

```python
#this is a simple grab and download lightcurve block
#grabs targets from the Mega_Target list and then searches lightcurves by K2 from those and downloads them if found
file_path = 'ASTR502_Mega_Target_List.csv'
download_dir = '/Users/iansterrett/Downloads/'
data = pd.read_csv(file_path)
tic_names = data['tic_id']
k2_target_list = []
for target in tqdm(tic_names):
    try:
        search_result = lk.search_lightcurve(f'{target}', mission='K2')
        if len(search_result) == 0:
            continue
        search_result.download_all(quality_bitmask='default', download_dir=download_dir, cutout_size=None)
        k2_target_list.append(target)
    except (LightkurveError, TimeoutError) as e:
        continue
```

```python
#this is to create a boolean list of which targets have and have not been observed by K2 to append to the master CSV
k2_target_list = np.array(k2_target_list)
k2_target_list = np.array(['TIC 201248411'])
tic_names = np.array(tic_names)
data["k2_observed"] = data["tic_id"].astype(str).isin(
    set(map(str, k2_target_list))
)
data["k2_status"] = data["k2_observed"].map(
    {True: "Observed", False: "Not observed"}
)
print(data["k2_status"])
```
